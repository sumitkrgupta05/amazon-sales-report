```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Cleaning

```python
# Reading the csv file
df=pd.read_csv("Amazon Sale Report.csv")
df.head(3)
```

```python
# No of rows & cols
df.shape
```

```python
# Returns all the information about the rows and columns of the data
df.info()
```

```python
# Deleting the empty columns
df.drop(['New','PendingS'],axis=1,inplace=True)
```

```python
# Checking the number of null values
# pd.isnull(df)

# Return number in each column which is empty
pd.isnull(df).sum()
```

```python
# Deleting all the null values
df.dropna(inplace=True)
# inplace=True to save the changes else it will just update it
```

```python
df.shape
```

```python
df.columns
```

```python
# changing column datatype
df['ship-postal-code']=df['ship-postal-code'].astype(int)
```

```python
df['ship-postal-code'].dtype
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Date'].dtype
```

```python
# Renaming column name
df.rename(columns={'Qty':'Quantity'})
```

```python
# To get detail but returns in int only
df.describe()
```

```python
# Return for onject datatype
df.describe(include='object')
```

```python
# To get details for targeted coulmns
df[['Qty','Amount']].describe()
```

## Exploratory Data Analysis

```python
df.columns
```

```python
 ax=sns.countplot(x='Size',data=df)
```

```python
# creating a count label using loop
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

##### From Graph we can see that most of the people buy M-Size

## Group By()

#### The groupby() function in pandas is used to group data based on one or more columns in a DataFrame

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
s_qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=s_qty)
plt.show()
```

#### Note: From the above graph we can see that most of the Quantity buys M-Size

## Courier Status

```python
sns.countplot(data=df,x='Courier Status',hue='Status')
```

```python
# Scale graph (Bda size ka graph)
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='Courier Status',hue='Status')
plt.show()
```

#### Note: From the above graph we can see that the majority of the orders are shipped though the courier

```python
# Histogram
df['Size'].hist()
```

```python
# change cateegory fron object to string and store in var
df['Category']=df['Category'].astype(str)
column_data=df['Category'];
plt.figure(figsize=(10,5))
#bin => same gap btwn all the row
plt.hist(column_data, bins=20, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

##### Note: From above graph we can see that most of the buyers are T-Shirt

```python
# Checking B2B data using pie chart
B2B_check=df['B2B'].value_counts()
# pie chart plotting 
plt.pie(B2B_check, labels=B2B_check, autopct="%1.1f%%")
# plt.axis('equal')
plt.show()
```

```python
# Checking B2B data using pie chart
B2B_check=df['B2B'].value_counts()
# pie chart plotting 
plt.pie(B2B_check, labels=B2B_check.index, autopct="%1.1f%%")
# plt.axis('equal')
plt.show()
```

##### Note: From the above chart we can see that maximum i.e, 99.2% of the buyers are retailers and 0.8% are the B2B buyers

```python
# Preparing the scatter plot
x_data=df['Category']
y_data=df['Size']
# Plot the scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.xlabel('Size')
plt.title('Scatter Plot')
plt.show()
```

##### Note: From the graph we can see that what is out of stock in which size

```python
# Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top-10 States
top_state=df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_state.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

##### Note: From the above graph we can see that most of the buyer are from maharashtra state
